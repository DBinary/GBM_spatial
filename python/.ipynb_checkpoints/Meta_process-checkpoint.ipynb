```python
import os
import anndata as ad

section_ids = ['CHH','CYL','FLY','JLM','LA','LJZ','LSN','LWD','LWH','LYZ','LZT','WXF','ZJC','ZXM']
print(section_ids)
pathway = '/storage/zengjianyangLab/hulei/GBM/Raw_Data/spatial_decon_new'
Batch_list = []
for section_id in section_ids:
    print(section_id)
    adata = sc.read_h5ad(os.path.join(pathway,section_id+".h5ad"))
    adata.obsm['spatial']= adata.obs[['x', 'y']].values
    # filter cells
    adata = adata[adata.obs['nCount_Spatial']>1000]
    adata = adata[adata.obs['nFeature_Spatial']>200]
    adata = adata[adata.obs['nCount_Spatial']<adata.obs['nCount_Spatial'].quantile(0.99)]
    adata = adata[adata.obs['percent.mito']<50]

    
    # check whether the adata.X is sparse matrix
    if isinstance(adata.X, pd.DataFrame):
        adata.X = csr_matrix(adata.X)
    else:
        pass
        
    adata.var_names_make_unique(join="++")

    # make spot name unique
    adata.obs_names = [x+'_'+section_id for x in adata.obs_names]
    Batch_list.append(adata)
    del adata
    
adata_concat = ad.concat(Batch_list, label="slice_name", keys=section_ids)
adata_concat.obs["batch_name"] = adata_concat.obs["slice_name"].astype('category')
```
